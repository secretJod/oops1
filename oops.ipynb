```python
# problem 1
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposit of {amount} successful. New balance is {self.balance}"
        else:
            return "Invalid amount for deposit"

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            return f"Withdrawal of {amount} successful. New balance is {self.balance}"
        else:
            return "Insufficient funds for withdrawal or invalid amount"

# Example usage
account1 = BankAccount("12345", "John Doe", 1000)
print(account1.deposit(500))  # Output: Deposit of 500 successful. New balance is 1500
print(account1.withdraw(300))  # Output: Withdrawal of 300 successful. New balance is 1200
print(account1.withdraw(1500))  # Output: Insufficient funds for withdrawal or invalid amount
```

```python
#problem 2
class Employee:
    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self):
        # Assuming the bonus is 10% of the salary
        bonus = 0.1 * self.salary
        return bonus

    def display_details(self):
        print("Employee ID:", self.emp_id)
        print("Name:", self.name)
        print("Salary:", self.salary)

# Creating an instance of the Employee class
emp1 = Employee("E001", "Alice", 50000)

# Calling methods to calculate yearly bonus and display employee details
yearly_bonus = emp1.calculate_yearly_bonus()
emp1.display_details()
print("Yearly Bonus:", yearly_bonus)
```

```python
# problem 3
class VehicleRental:
    def __init__(self):
        self.available_vehicles = {'car': 10, 'bike': 15, 'scooter': 5}

    def rent_vehicle(self, vehicle_type):
        if vehicle_type in self.available_vehicles and self.available_vehicles[vehicle_type] > 0:
            self.available_vehicles[vehicle_type] -= 1
            print(f"{vehicle_type} rented successfully.")
        else:
            print(f"Sorry, {vehicle_type} is not available for rent.")

    def return_vehicle(self, vehicle_type):
        if vehicle_type in self.available_vehicles:
            self.available_vehicles[vehicle_type] += 1
            print(f"{vehicle_type} returned successfully.")

    def display_available_vehicles(self):
        print("Available Vehicles:")
        for vehicle_type, count in self.available_vehicles.items():
            print(f"{vehicle_type.capitalize()}: {count}")

# Creating an instance of the VehicleRental class
vehicle_rental = VehicleRental()

# Renting a vehicle
vehicle_rental.rent_vehicle('car')
vehicle_rental.rent_vehicle('bike')
vehicle_rental.rent_vehicle('scooter')

# Returning a vehicle
vehicle_rental.return_vehicle('car')

# Displaying available vehicles
vehicle_rental.display_available_vehicles()
```

```python
# problem 4
class Book:
    def __init__(self, title, author):
        self.title = title
        self.author = author
        self.available = True

    def __str__(self):
        return f"{self.title} by {self.author}"

class Library:
    def __init__(self):
        self.catalog = []

    def add_book(self, book):
        self.catalog.append(book)
        print(f"Added \"{book.title}\" to the library catalog.")

    def borrow_book(self, title):
        for book in self.catalog:
            if book.title == title and book.available:
                book.available = False
                print(f"You have borrowed \"{book.title}\".")
                return
        print(f"Sorry, \"{title}\" is either not available or does not exist in the library catalog.")

    def display_available_books(self):
        print("Available Books:")
        for book in self.catalog:
            if book.available:
                print(book)

# Creating an instance of the Library class
library = Library()

# Adding books to the library catalog
book1 = Book("Harry Potter and the Sorcerer's Stone", "J.K. Rowling")
book2 = Book("To Kill a Mockingbird", "Harper Lee")
book3 = Book("1984", "George Orwell")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

# Borrowing a book
library.borrow_book("To Kill a Mockingbird")

# Displaying available books
library.display_available_books()
```

```python
# problem 5
class Product:
    def __init__(self, product_id, name, quantity):
        self.product_id = product_id
        self.name = name
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        self.quantity = new_quantity
        print(f"Updated quantity for {self.name} to {self.quantity}.")

    def __str__(self):
        return f"Product ID: {self.product_id}, Name: {self.name}, Quantity: {self.quantity}"

class Inventory:
    def __init__(self):
        self.products = []

    def add_product(self, product):
        self.products.append(product)
        print(f"Added {product.name} to the inventory.")

    def update_product_quantity(self, product_id, new_quantity):
        for product in self.products:
            if product.product_id == product_id:
                product.update_quantity(new_quantity)
                return
        print(f"Product with ID {product_id} not found in the inventory.")

    def display_available_products(self):
        print("Available Products:")
        for product in self.products:
            print(product)

# Creating an instance of the Inventory class
inventory = Inventory()

# Adding products to the inventory
product1 = Product(1, "Laptop", 10)
product2 = Product(2, "Smartphone", 20)
product3 = Product(3, "Headphones", 15)

inventory.add_product(product1)
inventory.add_product(product2)
inventory.add_product(product3)

# Updating product quantity
inventory.update_product_quantity(2, 25)

# Displaying available products
inventory.display_available_products()
```

```python
class Shape:
    def __init__(self, length, width, height):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        raise NotImplementedError("Subclass must implement abstract method")

    def calculate_perimeter(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Rectangle(Shape):
    def calculate_area(self):
        return self.length * self.width

    def calculate_perimeter(self):
        return 2 * (self.length + self.width)

class Triangle(Shape):
    def calculate_area(self):
        # Using Heron's formula to calculate the area of the triangle
        s = (self.length + self.width + self.height) / 2
        return (s * (s - self.length) * (s - self.width) * (s - self.height)) ** 0.5

    def calculate_perimeter(self):
        return self.length + self.width + self.height

# Creating instances of the Rectangle and Triangle classes
rectangle = Rectangle(5, 10, 0)
triangle = Triangle(5, 6, 7)

# Calculating area and perimeter of the rectangle
rectangle_area = rectangle.calculate_area()
rectangle_perimeter = rectangle.calculate_perimeter()

# Calculating area and perimeter of the triangle
triangle_area = triangle.calculate_area()
triangle_perimeter = triangle.calculate_perimeter()

print("Rectangle - Area:", rectangle_area, "Perimeter:", rectangle_perimeter)
print("Triangle - Area:", triangle_area, "Perimeter:", triangle_perimeter)
```

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        self.posts.append(post_content)
        print(f"New post added to {self.username}'s profile.")

    def display_posts(self):
        print(f"Posts by {self.username}:")
        for post in self.posts:
            print(post)

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post]
        if matching_posts:
            print(f"Posts containing '{keyword}' by {self.username}:")
            for post in matching_posts:
                print(post)
        else:
            print(f"No posts containing '{keyword}' found in {self.username}'s profile.")

# Create an instance of the SocialMediaProfile class
profile = SocialMediaProfile("john_doe")

# Add posts to the profile
profile.add_post("Enjoying a beautiful day at the beach!")
profile.add_post("Just finished reading a great book. Highly recommend it.")

# Display all posts in the profile
profile.display_posts()

# Search for posts containing a specific keyword
profile.search_posts("beach")
```

```python
class ToDoList:
    def __init__(self):
        self.tasks = {}  # Using a dictionary to store tasks with their due dates

    def add_task(self, task, due_date):
        self.tasks[task] = due_date
        print(f"Task '{task}' added with due date {due_date}.")

    def mark_task_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]
            print(f"Task '{task}' marked as completed.")
        else:
            print(f"Task '{task}' not found in the ToDo list.")

    def display_pending_tasks(self):
        if self.tasks:
            print("Pending Tasks:")
            for task, due_date in self.tasks.items():
                print(f"Task: {task} - Due Date: {due_date}")
        else:
            print("No pending tasks in the ToDo list.")

# Create an instance of the ToDoList class
to_do_list = ToDoList()

# Add tasks to the ToDo list
to_do_list.add_task("Complete report", "2024-09-15")
to_do_list.add_task("Buy groceries", "2024-09-02")

# Display pending tasks
to_do_list.display_pending_tasks()

# Mark a task as completed
to_do_list.mark_task_completed("Buy groceries")

# Display pending tasks after marking one as completed
to_do_list.display_pending_tasks()
```
